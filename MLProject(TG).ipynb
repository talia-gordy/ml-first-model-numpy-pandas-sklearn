# **Talia Gordy's ML project**


# **load data**

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/talia-gordy/Downloaded-Datasets/refs/heads/main/spotify_data%20clean.csv')
df
```

# **Data Selection**


## Data Seperation as X and Y


```python
y = df['track_duration_min']
y
```

```python
x = df.drop('track_duration_min', axis = 1)
x
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

# **Model Building**

## Linear Regression

### **Training the model**

```python
from sklearn.linear_model import LinearRegression

non_numeric_cols = [
    'track_id', 'track_name', 'explicit', 'artist_name',
    'artist_genres', 'album_id', 'album_name', 'album_release_date', 'album_type'
]

x_train_numeric = x_train.drop(columns=non_numeric_cols, errors='ignore')
x_test_numeric = x_test.drop(columns=non_numeric_cols, errors='ignore')

lr = LinearRegression()
lr.fit(x_train_numeric, y_train)
```

### **Applying the model to make a prediction**

```python
y_lr_train_pred = lr.predict(x_train_numeric)
y_lr_test_pred = lr.predict(x_test_numeric)
```

```python
print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

### **Evaluate model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train):  ', lr_train_mse)
print('LR R2 (Train):  ', lr_train_r2)
print('LR MSE (Test):  ', lr_test_mse)
print('LR R2 (Test):  ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

## Random Forest

### **Training the model**

```python
from sklearn.ensemble import RandomForestRegressor

rf  = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train_numeric, y_train)
```

### **Applying the model to make a prediction**

```python
y_rf_train_pred = lr.predict(x_train_numeric)
y_rf_test_pred = lr.predict(x_test_numeric)
```

### **Evaluate model performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```
